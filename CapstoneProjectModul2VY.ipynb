**Pengujian Korelasi**  
Korelasi adalah tentang asosiasi dan tidak berimplikasi pada sebab akibat. Korelasi hanya mengukur seberapa kuat hubungan dan arah hubungan. Nilai korelasi berkisar pada -1 < r < 1, serta arah positif (+) dan arah negatif (-). Arah positif artinya bila satu variabel naik, maka variabel lain juga naik. Sedangkan, arah negatif memiliki hubungan yang berkebalikan, seperti hubungan populasi tikus di sawah dengan produksi padi.  

Hasil korelasi dapat dinilai dari standar.  
- Nilai 0 - 0.3     : berarti hubungan lemah
- Nilai 0.3 - 0.7   : berarti hubungannya sedang
- Nilai 0.7 - 1     : berarti kekuatan hubungannya kuat  

**Visualisasi Heatmap**  
Heatmap korelasi menampilkan matriks korelasi 2 dimensi antara dua variable atau lebih. Heatmap ini menggunakan sel berwarna atau dalam skala monokromatik. Nilai dimensi pertama muncul di baris tabel, sedangkan nilai dimensi kedua diwakili oleh kolom tabel. Nilai warna sel mengikuti besaran pengukuran nilai korelasi. Biasanya semakin berkorelasi warnanya semakin gelap atau bisa sebaliknya. Melalui plot ini, Anda dapat mengidentifikasi pola dan anomali dari korelasi dengan lebih cepat.

Dapat dilihat bahwa variabel 'Arrival Delay in Minutes' berkorelasi atau memiliki hubungan yang kuat dengan variabel 'Departure Delay in Minutes'. Dalam menangani data kosong atau *Missing Value*, ada 2 cara yang dapat dilakukan, yaitu:
- Pertama, menghapus baris/kolom yang berisi *missing value*. Tidak disarankan pemakaian langsung, karena dapat mempengaruhi distribusi dan hasil pengujian statistik.  
- Kedua, mengisi data yang kosong. Cara ini disarankan. Ada beberapa metode yang bisa digunakan untuk mengisi *missing value*, cara yang paling baik adalah dengan mengisi data yang hilang dengan nilai sebenarnya, atau sedekat mungkin dengan nilai asli. Dalam analisis ini, kita akan  mengisi *missing value* berdasarkan kolom lain yang secara statistik berkaitan dengan kolom yang memiliki *missing value*. Jika masih ada kolom yang tidak bisa diisi, barulah kita mengisi dengan angka *mean, median* atau *modus*. Menghapus data akan menjadi opsi terakhir.  

```python
# Melihat jumlah data yang hilang dalam dataset (Missing Value)
df.isna().sum()
```

Terdapat 310 data kosong di kolom `Arrival Delay in Minutes`. Mari kita lihat, pembagian data - data kosong di kolom 'Arrival Delay in Minutes' ini berdasarkan kolom 'satisfaction'.  

```python
# Jumlah pelanggan dengan data kosong di 'Arrival Delay in Minutes' berdasarkan 'satisfaction'
display(df[df['Arrival Delay in Minutes'].isna()]['satisfaction'].value_counts())
```

Terlihat bahwa pelanggan yang bersifat 'neutral or dissatisfied' sebanyak 182 tidak melakukan pengisian pada kolom 'Arrival Delay in Minutes'. Angka ini lebih banyak jika dibandingkan dengan pelanggan yang bersifat 'satisfied'. Oleh sebab itu, kita bisa berasumsi bahwa pelanggan yang bersifat netral atau tidak puas, tentunya enggan untuk mengisi lama waktu keterlambatan pada kedatangan. Untuk itu, kita akan mengisi data kosong di kolom 'Arrival Delay in Minutes' menggunakan `median` berdasarkan kategori pada kolom 'satisfaction'.

```python
# Mengisi Missing Value dengan Median
df['Arrival Delay in Minutes'].fillna(df.groupby('satisfaction')['Arrival Delay in Minutes'].transform('median'), inplace = True)
```

```python
# Jumlah data kosong setelah pengisian
df[['Arrival Delay in Minutes']].isna().sum()
```

Adapun alasan lain dari pemilihan nilai median untuk mengisi nilai pada data kosong adalah karena dilihat dari histogram di bawah ini, gambaran sebaran data pada kolom 'Arrival Delay in Minutes' memiliki distribusi tidak normal. Grafik ini menunjukan distribusi yang condong ke kanan. Sehingga, data kosong (Missing Value) yang terdapat pada kolom 'Arrival Delay in Minutes' di isi dengan nilai median.

**`Kelima`**, mengubah isi data pada kolom - kolom kategori dengan penilaian 0 : Tidak ingin mengisi, 1 : Sangat tidak puas, 2 : Tidak puas, 3 : Biasa saja, 4 : Puas, 5 : Sangat Puas.

```python
# Membangun dictionary pengertian nilai
Replace_Isi_Category = {0 : 'Tidak ingin mengisi',
                        1 : 'Sangat tidak puas',
                        2 : 'Tidak puas',
                        3 : 'Biasa saja',
                        4 : 'Puas',
                        5 : 'Sangat Puas'}

# Membangun variabel baru yang berisikan kolom - kolom terpilih untuk diubah isi datanya
Kolom_Terpilih = ['Inflight wifi service', 'Departure/Arrival time convenient', 'Ease of Online booking', 'Gate location', 
                  'Food and drink', 'Online boarding', 'Seat comfort', 'Inflight entertainment', 'On-board service', 'Leg room service',
                  'Baggage handling', 'Checkin service', 'Inflight service', 'Cleanliness']

# Rumusan pengubahan
for Nama in Kolom_Terpilih:
    df[Nama] = df[Nama].replace(Replace_Isi_Category)

df.head()
```

## **3.) Exploratory Data Analysis**  
Exploratory Data Analysis (EDA) merupakan proses dalam proyek Analisis Data yang dilakukan dengan cara menyelidiki dataset untuk menemukan pola, dan anomali, serta membentuk hipotesis berdasarkan pemahaman tentang dataset yang dimiliki. Exploratory Data Analysis biasanya dilakukan dengan pembuatan ringkasan statistik data numerik dalam dataset dan membuat berbagai macam visualisasi grafik untuk memahami data dengan lebih baik.  

Adapun beberapa visualisasi grafik yang akan digunakan pada analisis ini, yaitu:  
- Histogram adalah representasi grafis yang akurat dari distribusi data numerik dan perkiraan distribusi probabilitas dari variabel kuantitatif. Histogram hanya digunakan untuk memplot frekuensi kemunculan skor dalam kumpulan data kontinu yang telah dibagi ke dalam kelas-kelas, yang disebut bins. Pada analisis ini, histogram akan digunakan untuk menampilkan data sebaran serta melihat distribusi data pada kolom 'Age', 'Flight Distance', 'Departure Delay in Minutes', dan 'Arrival Delay in Minutes'.
- Countplot adalah grafik sederhana yang digunakan untuk menghitung jumlah data berdasarkan kategorinya. Pada analisis ini, countplot akan digunakan untuk menampilkan data pada 'Gender', 'Customer Type', 'Type of Travel', 'Class', dan 'satisfaction'.
- Pie Chart atau Diagram lingkaran adalah grafik melingkar yang dibagi menjadi beberapa bagian. Besarnya potongan dalam diagram menunjukkan besarnya proporsi dari jumlah total yang diwakilinya. Diagram lingkaran paling cocok dipakai untuk menggambarkan proporsi tiap - tiap bagian dari keseluruhan informasi tertentu. Oleh karena itu, pada Exploratory Data Analysis ini, juga akan menampilkan Pie Chart pada beberapa data agar lebih mudah untuk dipahami. Pie chart yang dibentuk akan menggunakan startangle = 90. ### Pola hubungan variabel 'satisfaction' dengan variabel 'Age'  
Bagaimana pengaruh variabel 'Age' terhadap kepuasan pelanggan (variabel 'satisfaction') ?

```python
# Membuat diagram distibusi sebaran data
plt.figure(figsize=(12,6))
sns.distplot(df[df['satisfaction']=='neutral or dissatisfied']['Age'],label=0)
sns.distplot(df[df['satisfaction']=='satisfied']['Age'],label=1)
plt.title('Distribusi Age Berdasarkan satisfaction')
plt.legend()
plt.show()
```

Dari grafik di atas, dapat dilihat bahwa pola sebaran data memiliki kecenderungan tidak berdistribusi normal. Namun, ini masih asumsi. Mari kita lakukan pengujian statistik untuk mendapatkan hasil yang lebih pasti.  

**Hypothesis:**  
Ho = Variabel 'Age' Berdistribusi Normal  
Ha = Variabel 'Age' Berdistribusi Tidak Normal  

```python
# Cek distribusi kolom Age
from scipy.stats import normaltest
stats, pval=normaltest(df['Age'])
if pval<=0.05:
    print('Distribusi Tidak Normal')    #Ha
else: 
    print('Distribusi normal')          #Ho 
```

Setelah melakukan pengujian distribusi, mari lakukan pengujian rata-rata dua populasi yang independen atau tidak berhubungan (Two Sample Independent T-Test). Namun, karena distribusi data tidak normal, maka kita gunakan pengujian statistik non parametrik (Mann Whitney).  

**Hypothesis:**  
Ho : Rata - rata Nilai pelanggan 'satisfied' = Rata - rata Nilai pelanggan 'neutral or dissatisfied'  
Ha : Rata - rata Nilai pelanggan 'satisfied' < Rata - rata Nilai pelanggan 'neutral or dissatisfied'

```python
# uji statistik, apakah ada perbedaan Age diantara pelanggan yang 'satisfied' dan 'neutral or dissatisfied'
from scipy.stats import mannwhitneyu
stats,pval = mannwhitneyu(*[df[df['satisfaction']==i]['Age'] for i in df['satisfaction'].unique()], alternative='less')
if pval <= 0.05:
    print('Tolak Ho')
else:
    print('Gagal Tolak Ho')
```

Ternyata, memang ada perbedaan yang signifikan antara pelanggan 'satisfied' dengan 'neutral or dissatisfied' berdasarkan variabel 'Age'.  

Untuk lebih jelas, mari buat grafik boxplot untuk melihat titik pusat data, kemiringan data, sejauh mana suatu distribusi data kita tersebar, serta outlier dalam data set.  

```python
sns.boxplot(y='Age', x='satisfaction', data=df)
plt.title('Boxplot Age berdasarkan satisfaction')
plt.show()
```

Dapat disimpulkan bahwa, sebagian besar pelanggan yang bersifat 'neutral or dissatisfied' berusia antara 25 - 50 tahun (usia rata-rata mereka sedikit di atas 30 tahun). Sedangkan, pelanggan yang bersifat 'satisfied' memiliki  berusia antara 35 - 50 tahun (dengan usia rata-rata di atas 40 tahun)

Setelah melakukan pengujian distribusi, mari lakukan pengujian rata-rata dua populasi yang independen atau tidak berhubungan (Two Sample Independent T-Test). Namun, karena distribusi data tidak normal, maka kita gunakan pengujian statistik non parametrik (Mann Whitney).  

**Hypothesis:**  
Ho : Rata - rata Nilai pelanggan 'satisfied' = Rata - rata Nilai pelanggan 'neutral or dissatisfied'  
Ha : Rata - rata Nilai pelanggan 'satisfied' < Rata - rata Nilai pelanggan 'neutral or dissatisfied'

```python
from scipy.stats import mannwhitneyu
stats,pval = mannwhitneyu(*[df[df['satisfaction']==i]['Arrival Delay in Minutes'] for i in df['satisfaction'].unique()], alternative='less')
if pval <= 0.05:
    print('Tolak Ho')
else:
    print('Gagal Tolak Ho')
```

Hasilnya, Tidak ada perbedaan yang signifikan antara pelanggan 'satisfied' dengan 'neutral or dissatisfied' berdasarkan variabel 'Arrival Delay in Minutes'.
